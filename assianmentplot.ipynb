```python
import matplotlib.pyplot as plt

x = [0, 5, 9, 10, 15, 20, 25] 

y = [0, 1, 2, 3, 4, 5, 6]
# Create a line plot
plt.plot(x, y)
# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")
# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000, 36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000])

# Plotting
plt.subplot(2, 2, 1)
plt.plot(months, electronics_sales, marker='o', color='b')
plt.title('Electronics Sales')
plt.ylabel('Sales ($)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(months, clothing_sales, marker='o', color='g')
plt.title('Clothing Sales')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(months, home_garden_sales, marker='o', color='r')
plt.title('Home & Garden Sales')
plt.ylabel('Sales ($)')
plt.xlabel('Month')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(months, sports_outdoors_sales, marker='o', color='m')
plt.title('Sports & Outdoors Sales')
plt.xlabel('Month')
plt.grid(True)

plt.tight_layout()
plt.show()
```
